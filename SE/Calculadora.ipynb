```python
from tkinter import *
from math import * 
import parser

root = Tk()
root.title("Calculadora")


display = Entry(root)
display.grid(row=1, columnspan=6, sticky=W+E)

i = 0

def numeros(n):
    global i
    display.insert(i, n)
    i += 1

def operacion(operator):
    global i
    longitud_op = len(operator)
    display.insert(i, operator)
    i+=longitud_op

def calcular():
    display_state = display.get()
    try:
        math_expression = parser.expr(display_state).compile()
        result = eval(math_expression)
        clear_display()
        display.insert(0, result)
    except Exception:
        clear_display()
        display.insert(0, 'Error')

def clear_display():
    display.delete(0, END)


def undo():
    display_state = display.get()
    if len(display_state):
        display_new_state = display_state[:-1]
        clear_display()
        display.insert(0, display_new_state)
    else:
        clear_display()
        display.insert(0, 'Error')

#Numeros
Button(root, text="1", command=lambda: numeros(1), height=2, width=5).grid(row=2, column=0)
Button(root, text="2", command=lambda: numeros(2), height=2, width=5).grid(row=2, column=1)
Button(root, text="3", command=lambda: numeros(3), height=2, width=5).grid(row=2, column=2)

Button(root, text="4", command=lambda: numeros(4), height=2, width=5).grid(row=3, column=0)
Button(root, text="5", command=lambda: numeros(5), height=2, width=5).grid(row=3, column=1)
Button(root, text="6", command=lambda: numeros(6), height=2, width=5).grid(row=3, column=2)

Button(root, text="7", command=lambda: numeros(7), height=2, width=5).grid(row=4, column=0)
Button(root, text="8", command=lambda: numeros(8), height=2, width=5).grid(row=4, column=1)
Button(root, text="9", command=lambda: numeros(9), height=2, width=5).grid(row=4, column=2)

# Bottom Buttons
Button(root, text="AC", command=lambda: clear_display(), height=2, width=5).grid(row=5, column=0)
Button(root, text="0", command=lambda: numeros(0), height=2, width=5).grid(row=5, column=1)
Button(root, text="%", command=lambda: operacion("%"), height=2, width=5).grid(row=5, column=2)

Button(root, text="+", command=lambda: operacion("+"), height=2, width=5).grid(row=2, column=3)
Button(root, text="-", command=lambda: operacion("-"), height=2, width=5).grid(row=3, column=3, sticky=W+E)
Button(root, text="*", command=lambda: operacion("*"), height=2, width=5).grid(row=4, column=3, sticky=W+E)
Button(root, text="/", command=lambda: operacion("/"), height=2, width=5).grid(row=5, column=3, sticky=W+E)

# More Math Operators
Button(root, text="‚üµ", command=lambda: undo(), height=2, width=5).grid(row=2, column=4, sticky=W+E, columnspan=2)
Button(root, text="exp", command=lambda: operacion("**"), height=2, width=5).grid(row=3, column=4)
Button(root, text="^2", command=lambda: operacion("**2"), height=2, width=5).grid(row=3, column=5)
Button(root, text="(", command=lambda: operacion("("), height=2, width=5).grid(row=4, column=4,sticky=W+E)
Button(root, text=")", command=lambda:operacion(")"), height=2, width=5).grid(row=4, column=5, sticky=W+E)
Button(root, text="=", command=lambda: calcular(), height=2, width=5).grid(row=5, column=4, sticky=W+E, columnspan=2)

root.mainloop()
```
